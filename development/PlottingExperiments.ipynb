```python
from random import randrange
from nupic.research.spatial_pooler import SpatialPooler
```

```python
def generateInput(inputArray):
  inputArray[0:] = 0
  for i in range(inputArray.size):
    inputArray[i] = randrange(2)
```

```python
sp = SpatialPooler(inputDimensions=[20, 20],
                   columnDimensions=[15, 15],
                   potentialRadius=4,
                   potentialPct=0.5,
                   synPermInactiveDec=0.1,
                   synPermActiveInc=0.1,
                   synPermConnected=0.1,
                   localAreaDensity=0.1,
                   numActiveColumnsPerInhArea=-1,
                   globalInhibition=False)
```

```python
inputArray = np.zeros(sp.getNumInputs())
activeArray = np.zeros(sp.getNumColumns())
```

```python
Patcher().patchSP(sp)
```

```python
for i in range(50):
    generateInput(inputArray)
    sp.compute(inputArray, True, activeArray)
print "done"
```

```python
#import numpy
#import matplotlib.pyplot as plt
import matplotlib.ticker

def arrayPlot(array):
    (figure,axes) = plt.subplots()
    
    cAxes = axes.matshow(array,cmap='binary')
    
    figure.colorbar(cAxes)
    
    loc = matplotlib.ticker.IndexLocator(1,0)
    
    axes.xaxis.set_minor_locator(loc)
    axes.yaxis.set_minor_locator(loc)
       
    axes.grid(True,axis='both',which='minor',linestyle='solid',color=(0.7,0.7,0.7))
    
    #figure.set_size_inches(50,50)
```

# Hinton Plot - copied from matplotlib examples

```python
def hinton(matrix, max_weight=None, ax=None):
    """Draw Hinton diagram for visualizing a weight matrix."""
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2**np.ceil(np.log(np.abs(matrix).max())/np.log(2))

    ax.patch.set_facecolor('gray')
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x,y),w in np.ndenumerate(matrix):
        color = 'white' if w > 0 else 'black'
        size = np.sqrt(np.abs(w))
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    ax.autoscale_view()
    ax.invert_yaxis()
```

```python
hinton(np.random.rand(20, 20) - 0.5)
```
