# Introduction / Simple Demo

The idea of this project is to make it faster and easier to explore nupic.

This project provides two things:
 - the ability to save out data from NuPIC models as they run (and load it back in after).
 - plotting functions to visualize this data

These features are explained below in the context of a simple experiment. Run each cell in order using ```Shift-Enter```.

# A Simple Experiment

The plan is to feed a a spatial pooler with a repeating pattern, and then visualize what happens.  
If your running this notebook then run the cell below first to stop IPython automatically saving any changes you make.

````python
%autosave 0
````

## Get Some Input

First step is to get some input to feed the spatial pooler. This step just uses numpy.

````python
import numpy as np
````

````python
testInput = np.zeros(10)
for i in xrange(0,4):
    testInput[i]=1
````

````python
print testInput
````

A 1D array with a few consequtive bits active. We can then slide the on bits around to get a repeating pattern of inputs.  
However this is very small. We'll scale up to 200 bits to make it a bit more realistic.

````python
def getInput(iteration):
    array = np.zeros(200)
    position = iteration*4
    for i in range(0+position, 20+position):
        array[i%200] = 1 #this modulo is to make the on bits loop round nicely
    return array
````

````python
print getInput(0)
````

````python
print getInput(25)
````

````python
print getInput(48)
````

## Set Up The Model

Set up a spatial pooler from nupic.  

````python
from nupic.research.spatial_pooler import SpatialPooler
````

````python
sp = SpatialPooler(inputDimensions=[1, 200], columnDimensions=[1, 200],
                   potentialRadius=10, potentialPct=0.5,
                   synPermInactiveDec=0.1, synPermActiveInc=0.1,
                   synPermConnected=0.1, localAreaDensity=0.1,
                   numActiveColumnsPerInhArea=-1,globalInhibition=True)
````

## Using The Patcher

This is where nitpick starts to come in. In order to visualize what happens we need to record what happened.  
The Patcher object handles this.

```Patch()``` sets up a patcher object that handles saving data (with no argument this saves to the default directory).  
```.patchSP()``` modifies that spatial pooler so that it saves some of its state on every run.

````python
from nitpick.fileIO import Patcher
````

````python
Patcher().patchSP(sp)
````

Now everytime sp.compute() is called its state will be saved to '/tmp/cerebro2/model'

----------------------------
This saves all the data in the default directory ```/tmp/cerebro2/model```  
A different directory can be used by calling:

    Patcher('/my/special/directory/').patchSP(sp)

The patcher can also patch temporal pooler objects and opf models:
    
    Patcher().patchTP(...)
    Patcher().patchCLAModel(...)

## Run The Model

Feed in the input into the spatial pooler, 300 iterations total. 
This will mean the input loops around 6 times

````python
for i in xrange(300):
    inputArray = getInput(i)
    activeArray = np.zeros(sp.getNumColumns())
    sp.compute(inputArray, True, activeArray) #args are the 'input', 'whether to enable learning', and 'an array to write the output to'.
    
print 'done'
````

The data has been fed through the model. But what happened?  
This is were nitpick becomes very useful.

## Using the History Object

The past state of the model can be retrived anytime using the history object.

````python
sp.history.dimensions()
````

````python
print sp.history.activeColumns()[:10]
````

When the patcher runs, it automatically adds a history object to the patched thing, in this case a spatial pooler.  

Any saved state can be loaded by calling the methods on the history object. Use autcomplete (```Tab``` in IPython) to see them all.  

The data is returned as lists, with one item for each timestep.

------------

The history object can also be used independently for when you have saved data, but not the code that generated it.  
It works much like the Patcher object from above.

````python
from nitpick.fileIO import History
````

````python
History().dimensions()
````

This loads the data from the default directory ```/tmp/cerebro2/model```   
A different directory can be used by calling:

    History('/my/special/directory/').dimensions()

With just the raw data you can start to do some interesting things.

````python
sp.history.proximalSynapses()[0][:10]
````

Like calculate the mean permanence...

````python
meanPerms = [np.asarray(synapses)[:,2].mean() for synapses in sp.history.proximalSynapses()]
np.array(meanPerms).mean()
````

But more interestingly you can use nitpick's plots to plot the data.

## Using the Plotting Functions

All the plotting functions are in the plots module.  
This has been designed for autocomplete so typying ```plots.``` and then the ```Tab``` key will show all the plot function.

````python
import nitpick.plots as plots
````

````python
#this is required in IPython notebooks to show plots inline
%matplotlib inline
````

````python
plots.timeOrderedSDRPlot(sp.history.activeColumns())
#plt.gcf().set_size_inches(19,40)
````

This figure shows succesive SDR's. Each horizontal line represents one bit in the SDR (in this case bits are columns) and the dots show when that bit (column) was active, with time along the bottom axis.  
So if you draw a vertical line at some point in time, then all the dots it passes through show the active bits at that timestep. Note that all topology has been removed from this. The horizontal lines (ie columns) have been changed from their original ordering. They're now ordered according to when the bit was first active.

````python
plots.densityOrderedSDRPlot(sp.history.activeColumns())
````

This plot is the exact same as above but with the horizontal lines in a different order.  
They're ordered with the most frequently active bits at the bottom. The numbers on the left show how many times the bits in the band above were active.

All these plots are written in matplotlib, so they can be modified easily.  
Each function returns a tuple: ```("matplotlib figure", "list of axes")```
Below the figure is a bit small, so we can make it bigger.

````python
plots.differenceBarChart(sp.history.activeColumns()[:150],sp.history.dimensions())
````

````python
import matplotlib.pyplot as plt
(fig,ax) = plots.differenceBarChart(sp.history.activeColumns()[:150],sp.history.dimensions())
fig.set_size_inches(20,5)
````

The plot above shows the difference between succesive SDR's. Each bar sits in between two timesteps, showing the comparisons between them.  
The dark orange bar is the number of active bits that are the same between the two SDR's  
The dark green bar is the number of bits that became active (that weren't active before).  
The green bar is the number of bits that turned off (that were active before).  

More usefully:  
Dark two bars together show the number of active bits (at the later of the two timesteps)  
Top two green bars show the hamming distance between the two timesteps.

Also the percentage of the total SDR is on the right.  

The plot below converts an input sparse array into a full array and plots it.  
One of the useful features is the reshape option which will try and reshape the array into a better shape for viewing. You can also pass an aspect ratio to it to get a specific shape.

This shows the active columns at the 250th timstep.

````python
plots.sparseArrayPlot(sp.history.activeColumns()[250],sp.history.dimensions(),reshape=True)
````

This shows every permanence value at the 299 timestep.

````python
plots.permanencePlot(sp.history.proximalSynapses()[299],200,200)
````

And this shows all the permanence values across time, but only for column 100.

````python
plots.permanenceEvolutionPlot(sp.history.proximalSynapses(),100)
plt.gcf().set_size_inches(19,5)
````

## Going Further

Its also possible to plots things as you need them.  
For example to investigate the spread of permenences for a particular column:

````python
#get all the permanences for the column '43'
perms = []
for timestep in sp.history.proximalSynapses():
    permanences = [synapse[2] for synapse in timestep if synapse[0]==43]
    perms = perms+permanences
````

````python
perms[:10]
````

A histogram of the permanences of synapses connected to column 43 (over the whole time period):

````python
import matplotlib.pyplot as plt
````

````python
plt.hist(perms,bins=50,color=(0.3,0.4,0.8));
````
