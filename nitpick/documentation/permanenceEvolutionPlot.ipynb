# Permanence Evolution Plot

This plots the permanence of all the synapses connected to a particular column or cell over time.  
The first argument is a list of all synapses. The last argument is the index of the column or cell for which to plot permanences for.

### Get some input

```python
from nitpick.fileIO import History
h = History('/home/ruaridh/nitpick/nitpick/documentation/exampleData')
```

```python
perms = h.proximalSynapses()
```

### Draw The Plot

```python
from nitpick import plots
plots.permanenceEvolutionPlot(perms, 20)
```

### Implementation Details

 - All the data processing is done in pure python and is very slow and memory intensive.
 - The synapses saved by the patcher don't save any synapse with permanence 0. This was left over from an optimisation in Cerebro2. A lot of the code in this function is putting the 0 perm syanpses back in.
 - With a new patcher all will be fine
