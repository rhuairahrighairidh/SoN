# Permanence Plot

This plots an adjacency matrix for a list of synapses formatted as [source cell, end cell, permanence].  
The last two arguments are the 'total number of source cells' and 'total number of end cells'

### Get some input

```python
from nitpick.fileIO import History
h = History('/home/ruaridh/nitpick/nitpick/documentation/exampleData')
```

```python
timeStep50 = h.proximalSynapses()[50]
```

```python
timeStep50[:10]
```

### Draw The Plot

```python
h.dimensions()
```

```python
h.inputDimensions()
```

```python
from nitpick import plots
numColumns = 20*20
numInputs = 25*25
plots.permanencePlot(timeStep50, numColumns, numInputs)
```
