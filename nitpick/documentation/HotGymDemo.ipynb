The function to convert a dicionary to HTML was taken from Stack Overflow and isn't actually part of IPython.  
Currently its neat but it would be nice to incorporate something similar into nitpick.

````python
from IPython import display
def convertToHTML(dictObj, indent=0):
    p=[]
    p.append('<ul>\n')
    for k,v in dictObj.iteritems():
        if isinstance(v, dict):
            p.append('<li><b>'+str(k)+ '</b>:           ')
            p.append(convertToHTML(v))
            p.append('</li>')
        else:
            p.append('<li><b>'+ str(k)+ '</b>:        '+ str(v)+ '</li>')
    p.append('</ul>\n')
    return '\n'.join(p)
````

## Run Model

With the data loaded and a model set up we can now run the data through the model. This is were nitpick starts being useful.  
The aim is to be able to see what happened during the run. In order to do this, the state of the model at each timestep must be saved. This is done with the Patcher (which was stolen from Chetan's [Cerebro2](https://github.com/numenta/nupic.cerebro2)).

````python
from nitpick.fileIO import Patcher
````

````python
p=Patcher().patchCLAModel(m)
````

The above line setup the model to save out a load of its internal variables everytime the model.run() method is run.

Run the data through the model:  
Currently this is a bit messy because it has has the beginnings of a live updating plot. This plot code will hopefully incorporated into nitpick soon.  

````python
#Set up the plot
from nitpick.plots.displayFigureNow import displayFigureNow
from nupic.data.inference_shifter import InferenceShifter
shifter = InferenceShifter()
(fig,ax)=plt.subplots()
fig.set_size_inches(19,5)
results = []

for i in range(250):
    #Pull one piece from the hotgym data
    inputField = data.iloc[i].to_dict()
    #and convert into into a format nupic accepts
    inputField['timestamp'] = inputField['timestamp'].to_datetime()
    inputField['kw_energy_consumption'] = np.asscalar(inputField['kw_energy_consumption'])
    
    #Run the model
    result = m.run(inputField)
    
    #plot the results
    result = shifter.shift(result)
    results.append([result.rawInput['kw_energy_consumption']]+result.inferences['multiStepBestPredictions'].values())
    ax.clear()
    lines = ax.plot(results)
    displayFigureNow(fig)

#close the plot to stop IPython displaying it twice
plt.close()
````

### Going Further: Looking at the temporal pooler synapses.

The above plots are about as far as nitpick is at the moment. There isn't many, but each plot is pretty general so you can do a lot with them.  
Below is a bit of an experiment, using data the patcher doesn't (yet) save. The idea is that connections in the TP form a network, so what does that network look like?

````python
tp = m._getTPRegion().getSelf()._tfdr

##There must be beter ways of doing this.
def getAdjacencyMatrix(tp):
    """
    Return a binary adjacency matrix for the synapses in the temporal pooler object, tp.
    columns - cell that own the syanpses
    rows - cells that the synapses are connected to
    """
    adjacencyMatrix = np.zeros([tp.numberOfCols*tp.cellsPerColumn,tp.numberOfCols*tp.cellsPerColumn])

    for col in xrange(tp.numberOfCols):
        for c in xrange(tp.cellsPerColumn):
            cell = tp.cells4.getCell(col,c)
            for s in xrange(cell.nSegments()):
                seg = cell.getSegment(s)
                for syn in xrange(cell.nSynapses()):
                    try:
                        srcCell = seg.getSrcCellIdx(syn)
                        adjacencyMatrix[srcCell][col*tp.cellsPerColumn+c] = 1#seg.getPermanence(syn)
                    except:
                        pass
    return adjacencyMatrix

def getPermAdjancencyMatrix(tp):
    """
    Return a real valued adjacency matrix for the synapses in the temporal pooler object, tp, were each value shows the permanence of the syanapse
    columns - cell that own the syanpses
    rows - cells that the synapses are connected to
    """
    adjacencyMatrix = np.zeros([tp.numberOfCols*tp.cellsPerColumn,tp.numberOfCols*tp.cellsPerColumn])

    for col in xrange(tp.numberOfCols):
        for c in xrange(tp.cellsPerColumn):
            cell = tp.cells4.getCell(col,c)
            for s in xrange(cell.nSegments()):
                seg = cell.getSegment(s)
                for syn in xrange(cell.nSynapses()):
                    try:
                        srcCell = seg.getSrcCellIdx(syn)
                        adjacencyMatrix[srcCell][col*tp.cellsPerColumn+c] = seg.getPermanence(syn)
                    except:
                        pass
    return adjacencyMatrix
````

````python
adjM=getAdjacencyMatrix(tp)
````

This is the adjacency matric for the temporal pooler synapes. Each side is 1024 (the number of cells). Each synapse, of any permanence, is shown by a black dot. The indexes on the right are the indexes of the cells that own the synapses. The top indexes are of the cells the synapses connect to.

````python
plt.matshow(adjM,cmap='binary')
plt.gcf().set_size_inches(20,20)
````
